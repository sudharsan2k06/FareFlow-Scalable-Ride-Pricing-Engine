```python
import pandas as pd,numpy as np
import warnings
warnings.filterwarnings('ignore')
combine_df = pd.read_parquet('../data/sampled/combined_sample.parquet')
print("null check")
print(combine_df.isnull().sum())
print("duplicate check")
print(combine_df[combine_df.duplicated()])
print("data types")
print(combine_df.dtypes)
print("target statistics")

## this section will remove any edges cases in the target 

combine_df= combine_df[combine_df['base_passenger_fare'] > 0]
combine_df = combine_df[combine_df['trip_miles'] > 0]
combine_df = combine_df[combine_df['trip_time'] > 0]

# Remove extreme outliers (optional but recommended)
fare_upper = combine_df['base_passenger_fare'].quantile(0.99)
combine_df = combine_df[combine_df['base_passenger_fare'] <= fare_upper]

print(f"Rows after cleanup: {combine_df.shape[0]}")
```
