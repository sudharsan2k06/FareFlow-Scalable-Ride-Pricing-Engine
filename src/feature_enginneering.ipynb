```python
import pandas as pd
```

```python
input_file = "../data/processed/clean_split_10.parquet"
output_file = "../data/feature_engineered/features_split_10.parquet"
```

```python
# load cleaned split
df = pd.read_parquet(input_file)

# -------- Feature Engineering --------

df["trip_minutes"] = df["trip_time"] / 60

df["trip_speed"] = df["trip_miles"] / (df["trip_minutes"] + 0.01)

df["pickup_hour"] = df["pickup_datetime"].dt.hour
df["pickup_dayofweek"] = df["pickup_datetime"].dt.dayofweek
df["is_weekend"] = df["pickup_dayofweek"].isin([5,6]).astype(int)

# -------- Drop raw datetime columns --------

df = df.drop(columns=[
    "request_datetime",
    "on_scene_datetime",
    "pickup_datetime",
    "dropoff_datetime"
])

# save engineered dataset
df.to_parquet(output_file, index=False)
```
